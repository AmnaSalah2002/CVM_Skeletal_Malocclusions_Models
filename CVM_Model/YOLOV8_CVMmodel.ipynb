```python
from ultralytics import YOLO
```

```python
%cd /content/drive/MyDrive/CVMYOLOV8DataSet
```

```python
model = YOLO("yolov8m.pt")
```

```python
model.train(data='/content/drive/MyDrive/CVMYOLOV8DataSet/dataset.yaml', epochs=300, imgsz=640, batch=16, optimizer="AdamW", patience=30,
    lr0=0.0005, lrf=0.01, dropout=0.1, cos_lr=True, kobj=7.0, box=10.0, cls=1.0, dfl=2.0, weight_decay=0.0001, momentum=0.85,
    warmup_epochs=3, warmup_momentum=0.85, warmup_bias_lr=0.1, conf=0.4, iou=0.65, hsv_s=0.01, hsv_v=0.01, degrees=1.0, translate=0.01,
    scale=0.01, fliplr=0.0, crop_fraction=0.0, mosaic=0.0, mixup=0.0 )
```

```python
best_model = YOLO('/content/drive/MyDrive/CVMYOLOV8DataSet/runs/detect/train/weights/best.pt')

metrics = best_model.val(
    data='/content/drive/MyDrive/CVMYOLOV8DataSet/dataset.yaml',
    split='test')

print(f"Mean Average Precision (mAP50-95): {metrics.box.map}")
print(f"mAP50: {metrics.box.map50}")
```
